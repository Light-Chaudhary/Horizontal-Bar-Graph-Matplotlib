```python
# Bar Chart Comparison
# Instructions:

# Use the same players and years: Kohli, Rohit, Dhawan, and years 2015–2020.

# Create a grouped bar chart where bars for each player appear side by side for each year.

# Use different colors for each player.

# Add:

# Title

# X and Y axis labels

# Legend

# Grid (optional)

# Make sure the years appear correctly on the x-axis and bars don’t overlap.

import matplotlib.pyplot as plt
import numpy as np
years = [2015, 2016, 2017, 2018, 2019, 2020]

kohli_runs = [850, 900, 1000, 1100, 950, 1050]
rohit_runs = [700, 850, 900, 950, 1000, 1100]
dhawan_runs = [650, 700, 750, 800, 850, 900]

x = np.arange(len(years))
width = 0.25
plt.bar(x - width, kohli_runs, width = width , label = "Kohli Run")
plt.bar(x, rohit_runs, width = width, label = "Rohit Run")
plt.bar(x + width, dhawan_runs, width = width,  label = "Dhawan Run")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Bar Graph: Run comparison")
plt.grid(True)
plt.xticks(x, years)
plt.show()
```
